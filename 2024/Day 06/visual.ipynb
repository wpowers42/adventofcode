```python
grid = df_input.pivot(index='y', columns='x', values='cell').to_numpy()


def print_grid(df_result, depth):
    guard = df_result
    x = guard.x[depth] - 1
    y = guard.y[depth] - 1
    dx = guard.dx[depth]
    dy = guard.dy[depth]
    char = 2 if dx == 0 and dy == -1 else 4 if dx == 0 and dy == 1 else 5 if dx == -1 and dy == 0 else 3
    cmap = {0: '.', 1: '#', 2: '^', 3: '>', 4: 'v', 5: '<'}
    o = grid.copy()
    o[y, x] = char

    # only print the rows 14 above and below the current position
    o = o[max(0, y - 10):min(130, y + 10), ]

    return '\n'.join([''.join(['{}'.format(cmap[item]) for item in row]) for row in o])
```

```python
for i in range(0, df_result.shape[0]):
    print(print_grid(df_result, i), end='\r')
    print(i)
    time.sleep(0.01)
    clear_output(False)
```
